```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df= sns.load_dataset("tips")
df.head()
```

```python
sns.distplot(df["total_bill"])
```

```python
sns.jointplot(x=df["total_bill"], y=df["tip"], data=df)
```

```python
sns.jointplot(x=df["total_bill"], y=df["tip"], data=df, kind="hex")
```

```python
sns.jointplot(x=df["total_bill"], y=df["tip"], data=df, kind="kde")
```

```python
sns.jointplot(x=df["total_bill"], y=df["tip"], data=df, kind="reg")
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="sex")
```

```python
sns.barplot(y=df["total_bill"], x=df["sex"], data= df)
```

```python
sns.countplot(df["sex"], data=df)
```

```python
sns.boxenplot(x=df["day"], y=df["total_bill"], data=df, hue="sex")
```

```python
sns.violinplot(x=df["day"], y=df["total_bill"], data=df, hue="sex")
```

```python
sns.heatmap(df.corr(), annot=True, linecolor='white', linewidths=3)
```
